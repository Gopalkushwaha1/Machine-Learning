# **Categorical Data**

**a. Countplot**

```python
sns.countplot(x='Survived', hue = 'Sex' , data=df)    # It count frequency of differnet cate...
plt.show()
```

```python
sns.countplot(x='Pclass', hue = 'Sex' , data = df)
```

```python
sns.countplot(x='Survived', data=df, palette='Set3')
```

```python
sns.countplot(x='Pclass' , data = df , palette='Set2')
```

```python
sns.countplot(x='Sex' , hue = 'Sex' , data=df)
```

**pie char**

```python
survived_counts = df['Survived'].value_counts()

# Plot pie
plt.pie(survived_counts, labels=survived_counts.index, autopct='%1.1f%%', colors=['red', 'green'])
plt.title('Survival Distribution')
plt.legend(['Not Survived', 'Survived'])
plt.show()
```

```python
pClass_count = df['Pclass'].value_counts()

# Plot pie graph
plt.pie(pClass_count , labels=pClass_count.index,autopct='%1.1f%%',colors=['red','green','yellow'])
plt.title('PClass Distribution')
plt.legend(['First Class' , 'Second Class' , 'Third Class'])
plt.show()
```
