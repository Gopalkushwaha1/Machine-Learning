# **1. Scatterplot (Numerical - Numerical)**

```python
sns.scatterplot(x='total_bill'  , y = 'tip' , data=tips, hue = 'sex' , style='smoker')
```

# **2. Bar Plot (Numerical-Categorical)**

```python
sns.barplot(x='Pclass',y='Age' , data = titanic , palette='Set2')
```

```python
sns.barplot(x='Pclass', y='Fare', data=titanic, hue='Sex', palette='Set2')
```

# **3. Box Plot (Numerical - Categorical)**

```python
sns.boxplot(x='Sex' , y = 'Age' , data = titanic)
```

# **4. Displot (Numerical - Categorical)**

```python
sns.histplot(data=titanic, x='Age', kde=True)
```

# **5. HeatMap(Categorical-Categorical)**

```python
pd.crosstab(titanic['Pclass'] , titanic['Survived'])
```

```python
table = pd.crosstab(titanic['Pclass'], titanic['Survived'])

# Plot the heatmap
sns.heatmap(table, annot=True, cmap='YlGnBu')
```

```python
titanic.groupby('Pclass')['Survived'].mean() * 100
```

# **6. ClusterMap (Categorical-Categorical)**

```python
pd.crosstab(titanic['SibSp'],titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```

# **7. Pairplot**

```python
sns.pairplot(iris)
```

# **8. LinePlot (Numerical-Numerical)**

```python
new = flights.groupby('year').sum(numeric_only=True).reset_index()
```

```python
sns.lineplot(x='year', y='passengers', data=new)
```
